```python
import xmljson
import xmltodict,json
from xml.etree.ElementTree import fromstring

file = "Dataset/arxiv/metadata_arxiv.tar.gz"
i=0
with open("Dataset/arxiv/arxiv_data.json","w") as jsonfile:
  with tarfile.open(file) as tf:
      for entry in tf:  # list each entry one by one
        i+=1
        #print(entry)
        fileobj = tf.extractfile(entry)
        if fileobj==None:
          continue        
        arxiv_row = fileobj.read()
        #print()
        arxiv_row = str(arxiv_row,encoding='UTF-8')
        arxiv_row = xmltodict.parse(arxiv_row)
        #print(json.dumps(arxiv_row))
        # convert this into one json file
        jsonfile.writelines(json.dumps(arxiv_row)+"\n")
        
        if i%1000==0:
          print(i)
        
        #if i > 10:
        #  break;
```

```python
def return_key(my_d,key):
  if key in my_d:
    return my_d[key]
  return None
```

```python
import json
import csv
file = "Dataset/arxiv/arxiv_data.json"
with open(file,"r") as file:
  main_csv =  open('Dataset/arxiv/arxiv_main.csv', 'w')
  author_csv = open('Dataset/arxiv/arxiv_author.csv', 'w')
  abstract_csv = open('Dataset/arxiv/arxiv_abstract.csv', 'w')
  categories_csv = open('Dataset/arxiv/arxiv_categories.csv', 'w')
  main_writer = csv.writer(main_csv)
  author_writer = csv.writer(author_csv)
  abstract_writer = csv.writer(abstract_csv)
  categories_writer = csv.writer(categories_csv)
  for i,x  in enumerate(file):
    data = json.loads(x)
    data = data["arXiv"]
    #print(data)
    tid = data["id"]
    tcreated = data["created"]
    tupdated = return_key(data,"updated")
    ttitle = data["title"]
    tabstract = data["abstract"]
    
    tcategories=data["categories"]

    main_writer.writerow([tid,tcreated,tupdated])
    tauthors = data["authors"]["author"]
    if type(tauthors)==list:
      for author in tauthors:
        taff = return_key(author,"affiliation")        
        author_writer.writerow([tid,author["keyname"],return_key(author,"forenames"),taff])
    else:      
      taff = return_key(tauthors,"affiliation")
      author_writer.writerow([tid,tauthors["keyname"],return_key(tauthors,"forenames"),taff])
    
    for cat in tcategories.split(" "):
      #print(cat)
      main_cat = cat.split(".")
      categories_writer.writerow([tid,main_cat[0],cat])  
    
    abstract_writer.writerow([tid,ttitle,tabstract])
    
    """
    print(data["id"])
    print(data["categories"])
    print(data["created"])
    print(data["title"])
    print(data["authors"])
    print(data["abstract"])
    """
    #if i > 10:
    #  break
      
  main_csv.close()
  author_csv.close()
  abstract_csv.close()
  categories_csv.close()
```
