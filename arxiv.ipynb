```python
main_cat = main.merge(categories,how="inner",left_on="id",right_on="id")
```

```python
main_cat.sort_values("create_date",ascending=False).head()
```

```python
main_cat[(main_cat.cat=="math")&(main_cat.create_date.apply(lambda x:int(x[:4]))>=2012)]
```

```python
# look at papers greater than 2013
import random
train_set = pd.DataFrame()
test_set = pd.DataFrame()
for mc in my_cat:
    selected_cat = main_cat[(main_cat.cat==mc)&(main_cat.create_date.apply(lambda x:int(x[:4]))>=2013)].sort_values("create_date",ascending=False)
    total_rows = selected_cat.shape[0]
    if total_rows > 6000:
        ran_sample = random.sample(range(total_rows),6000)
        ran_train = ran_sample[:5000]
        ran_test = ran_sample[5000:]
        select_train = selected_cat.iloc[ran_train]
        train_set = train_set.append(select_train)
        select_test = selected_cat.iloc[ran_test]
        test_set = test_set.append(select_test)
    else:
        ran_sample = random.sample(range(total_rows),total_rows)
        total_sample = total_rows - int(0.2*total_rows)
        ran_train = ran_sample[:total_sample]
        ran_test = ran_sample[total_sample:]
        select_train = selected_cat.iloc[ran_train]
        train_set = train_set.append(select_train)
        select_test = selected_cat.iloc[ran_test]    
        test_set = test_set.append(select_test)    
```
